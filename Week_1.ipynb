```python
# Task no 3
# Add two numbers

number_1 = 3.0
number_2 = 1.5
sum = number_1 + number_2

print(f"Sum of {number_1} and {number_2} is {sum}")
```

```python
# Task no 4
# Add two numbers using input

number_1 = float(input("Enter your first Number: "))  # input() always returns a string, not a number.
number_2 = float(input("Enter your second Number: ")) # So if you enter: 3,4 ---> Python treats them like: "3" + "4" = "34"

sum = number_1 + number_2

print(f"sum of {number_1} and {number_2} is {sum}")
```

```python
# Task no 5
#find the square root of a number entered by the user

number = float(input("Enter a number: "))
number_sqrt = number ** 0.5

print(f"Square root of {number} is {number_sqrt}")
```
